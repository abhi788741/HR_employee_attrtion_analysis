# Feature Engineering -

```python
def age_category(age):
    if age>=20 and age<=30:
        return 'Young'
    elif age>30 and age<=40:
        return 'Adult'
    else:
        return 'Senior'
df["Age_Group"]=df["Age"].apply(age_category)
```

```python
def experience_level(years):
    if years==0 and years<=2:
        return 'Junior'
    elif years>2 and years<=5:
        return 'Mid_Level'
    elif years>5 and years<=10:
        return 'Senior'
    else:
        return 'Expert'
df["Experience_Level"]=df["Experience_Years"].apply(experience_level)
```
